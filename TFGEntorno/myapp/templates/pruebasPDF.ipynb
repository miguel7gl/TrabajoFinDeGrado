```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors
import os
from datetime import datetime
import pandas as pd

def get_user_email(filepath):
    with open(filepath, 'r') as file:
        line = file.readline().strip()
        email = line.split(':')[0]
    return email

def get_model(filepath):
    with open(filepath, 'r') as file:
        contenido = file.read().strip()
        elementos = contenido.split(':')
    objetivo, pc1, pc2, pc3, pc4, pc5, pc6 = elementos[:7]
    return round(float(objetivo),2), round(float(pc1),2), round(float(pc2),2), round(float(pc3),2), round(float(pc4),2), round(float(pc5),2), round(float(pc6),2)

def get_datos_from_csv(csv_path):
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        if not df.empty:
            cups = df.iloc[0, 1]  # Suponiendo que el número CUPS está en la primera fila, primera columna

            # Inicializar los vectores P1 a P6 como listas vacías
            P1, P2, P3, P4, P5, P6 = [], [], [], [], [], []

            # Iterar sobre las filas del DataFrame y llenar los vectores P1 a P6
            for index, row in df.iterrows():
                consumo = row['Consumo (kwh)']
                periodo = row['Periodo']
                if periodo == 'P1':
                    P1.append(consumo)
                elif periodo == 'P2':
                    P2.append(consumo)
                elif periodo == 'P3':
                    P3.append(consumo)
                elif periodo == 'P4':
                    P4.append(consumo)
                elif periodo == 'P5':
                    P5.append(consumo)
                elif periodo == 'P6':
                    P6.append(consumo)

            # Calcular máximo, mínimo y promedio para cada vector
            P1_max = max(P1) if P1 else None
            P1_min = min(P1) if P1 else None
            P1_avg = round(sum(P1) / len(P1), 2) if P1 else None

            P2_max = max(P2) if P2 else None
            P2_min = min(P2) if P2 else None
            P2_avg = round(sum(P2) / len(P2), 2) if P2 else None

            P3_max = max(P3) if P3 else None
            P3_min = min(P3) if P3 else None
            P3_avg = round(sum(P3) / len(P3), 2) if P3 else None

            P4_max = max(P4) if P4 else None
            P4_min = min(P4) if P4 else None
            P4_avg = round(sum(P4) / len(P4), 2) if P4 else None

            P5_max = max(P5) if P5 else None
            P5_min = min(P5) if P5 else None
            P5_avg = round(sum(P5) / len(P5), 2) if P5 else None

            P6_max = max(P6) if P6 else None
            P6_min = min(P6) if P6 else None
            P6_avg = round(sum(P6) / len(P6), 2) if P6 else None
            
            return cups, P1_max, P1_min, P1_avg, P2_max, P2_min, P2_avg, P3_max, P3_min, P3_avg, P4_max, P4_min, P4_avg, P5_max, P5_min, P5_avg, P6_max, P6_min, P6_avg

    return "No se encontró el número CUPS o el archivo CSV está vacío"

def create_pdf(filename, text, image_path, csv_path):
    # Crear un documento PDF con márgenes específicos
    doc = SimpleDocTemplate(filename, pagesize=letter, leftMargin=0.75*inch, rightMargin=0.75*inch, topMargin=0.75*inch, bottomMargin=0.75*inch)
    elements = []

    # Estilos de párrafo
    styles = getSampleStyleSheet()
    title_style = styles['Title']
    normal_style = styles['Normal']
    small_font_style = ParagraphStyle(
        'small',
        parent=normal_style,
        fontSize=8,  # Tamaño de fuente más pequeño
    )

    small_font_style = ParagraphStyle(
        'small',
        parent=normal_style,
        fontSize=8,  # Tamaño de fuente más pequeño
    )

    text_paragraph = Paragraph(text, styles['Normal'])

    # Ruta de la imagen del logo
    img_path_logo = "../static/logo.jpg"

    # Obtener datos del csv de partida
    cups, P1_max, P1_min, P1_avg, P2_max, P2_min, P2_avg, P3_max, P3_min, P3_avg, P4_max, P4_min, P4_avg, P5_max, P5_min, P5_avg, P6_max, P6_min, P6_avg = get_datos_from_csv(os.path.join("..", "..", csv_path))
    objetivo, pc1, pc2, pc3, pc4, pc5, pc6 = get_model("../../modelo.txt")

    ########################################### TÍTULO #################################################
    # Crear una tabla para la imagen y los títulos
    data = []
    if os.path.exists(img_path_logo):
        img_logo = Image(img_path_logo, 0.5*inch, 0.6*inch)
        img_logo.hAlign = 'LEFT'
        combined_paragraph = Paragraph('<b>POWER CALCULATOR<br/><font size="10">Informe</font></b>', title_style)
        data.append([img_logo, combined_paragraph])
    else:
        print(f"El logo '{img_path_logo}' no pudo ser encontrado.")
        combined_paragraph = Paragraph('<b>POWER CALCULATOR<br/><font size="10">Informe</font></b>', title_style)
        data.append(["", combined_paragraph])

    table = Table(data, colWidths=[0.5*inch, 6.5*inch])  # Ajustar colWidths para mejor alineación
    table.setStyle(TableStyle([
        ('VALIGN', (0, 0), (0, -1), 'MIDDLE'),  # Alinear verticalmente el logo en todas las filas
        ('VALIGN', (1, 0), (1, -1), 'MIDDLE'),  # Alinear verticalmente el texto en todas las filas
        ('ALIGN', (1, 0), (1, -1), 'CENTER'),
        ('BOTTOMPADDING', (0, 0), (-1, -1), 0),  # Reducir el espaciado inferior de todas las filas
        ('TOPPADDING', (0, 0), (-1, -1), 0)  # Reducir el espaciado superior de todas las filas
    ]))
    elements.append(table)
    elements.append(Spacer(1, 24))  # Espacio entre la tabla y "Datos del usuario"

    ########################################### DATOS DE USUARIO #################################################

    # Agregar texto "Datos del usuario" con fondo amarillo
    user_data_paragraph = Paragraph('<b>Datos del usuario</b>', styles['Normal'])
    user_data_table = Table([[user_data_paragraph]], colWidths=[7*inch])
    user_data_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), '#ffbf00'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(user_data_table)
    elements.append(Spacer(1, 12))

    # Agregar la fecha, hora y correo del usuario
    current_datetime = datetime.now()
    fecha_paragraph = Paragraph(f'Fecha: {current_datetime.strftime("%Y-%m-%d")}', styles['Normal'])
    hora_paragraph = Paragraph(f'Hora: {current_datetime.strftime("%H:%M:%S")}', styles['Normal'])
    email = get_user_email("../../login_data.txt")
    correo_paragraph = Paragraph(f'Correo: {email}', styles['Normal'])

    # Crear una tabla para alinear los datos del usuario
    user_info_table = Table(
        [[fecha_paragraph], [hora_paragraph], [correo_paragraph]],
        colWidths=[7*inch]
    )
    user_info_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(user_info_table)
    elements.append(Spacer(1, 12))

    ########################################### CARACTERÍSTICAS DEL SUMINISTRO #################################

    # Agregar texto "Características del suministro" con fondo amarillo
    characteristics_paragraph = Paragraph('<b>Características del suministro</b>', styles['Normal'])
    characteristics_table = Table([[characteristics_paragraph]], colWidths=[7*inch])
    characteristics_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), '#ffbf00'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(characteristics_table)
    elements.append(Spacer(1, 12))

    # Agregar los datos del suministro
    cups_paragraph = Paragraph(f'CUPS: {cups}', styles['Normal'])
    tariff_paragraph = Paragraph(f'Tarifa: 6.x TD', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[cups_paragraph], [tariff_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))

    ########################################### DATOS DE PARTIDA #################################################

    # Agregar texto "Datos de partida" con fondo amarillo
    starting_data_paragraph = Paragraph('<b>Datos de partida</b>', styles['Normal'])
    starting_data_table = Table([[starting_data_paragraph]], colWidths=[7*inch])
    starting_data_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), '#ffbf00'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(starting_data_table)
    elements.append(Spacer(1, 12))

    # Estilo para la tabla de datos de partida
    small_table_style = TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('GRID', (0, 0), (-1, -1), 1, colors.black),
        ('BACKGROUND', (0, 0), (-1, 0), colors.lightgrey),  # Color de fondo para la primera fila
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),  # Fuente en negrita para la primera fila
        ('FONTSIZE', (0, 0), (-1, -1), 8),  # Tamaño de fuente más pequeño para toda la tabla
    ])

    # Agregar una tabla para los valores máximos, mínimos y promedios
    data_table = [
        ['Periodo', 'P1', 'P2', 'P3', 'P4', 'P5', 'P6'],
        ['Valor máximo', P1_max, P2_max, P3_max, P4_max, P5_max, P6_max],
        ['Valor mínimo', P1_min, P2_min, P3_min, P4_min, P5_min, P6_min],
        ['Valor medio', P1_avg, P2_avg, P3_avg, P4_avg, P5_avg, P6_avg]
    ]

    table_data = Table(data_table, colWidths=[1*inch]*7)
    table_data.setStyle(small_table_style)
    elements.append(Spacer(1, 12))
    elements.append(table_data)
    elements.append(Spacer(1, 12))

    ########################################### GRÁFICA #######################################################

    # Obtener la ruta completa de la imagen
    img_path = os.path.join("..", "static", image_path)

    # Comprobar si la imagen existe y agregarla
    if os.path.exists(img_path):
        img = Image(img_path, 6*inch, 2*inch)
        elements.append(img)
        elements.append(Spacer(1, 12))
        print("Imagen agregada al PDF.")
    else:
        print(f"La imagen '{img_path}' no pudo ser encontrada.")


    ########################################### RESULTADOS DEL MODELO #################################

    # Agregar texto "Características del suministro" con fondo amarillo
    results_paragraph = Paragraph('<b>Resultados del modelo</b>', styles['Normal'])
    results_table = Table([[results_paragraph]], colWidths=[7*inch])
    results_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), '#ffbf00'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(results_table)
    elements.append(Spacer(1, 12))

    # Agregar los datos del suministro
    p1_paragraph = Paragraph(f'Periodo 1: {pc1}', styles['Normal'])
    p2_paragraph = Paragraph(f'Periodo 2: {pc2}', styles['Normal'])
    p3_paragraph = Paragraph(f'Periodo 3: {pc3}', styles['Normal'])
    p4_paragraph = Paragraph(f'Periodo 4: {pc4}', styles['Normal'])
    p5_paragraph = Paragraph(f'Periodo 5: {pc5}', styles['Normal'])
    p6_paragraph = Paragraph(f'Periodo 6: {pc6}', styles['Normal'])
    #objetivo_paragraph = Paragraph(f'Función objetivo: {objetivo}', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[p1_paragraph], [p2_paragraph], [p3_paragraph], [p4_paragraph], [p5_paragraph], [p6_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))

    ########################################### EXPLICACION ###################################################

    results_paragraph = Paragraph('<b>Explicación</b>', styles['Normal'])
    results_table = Table([[results_paragraph]], colWidths=[7*inch])
    results_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), '#ffbf00'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(results_table)
    elements.append(Spacer(1, 12))

    p6_paragraph = Paragraph(f'Para calcular la facturación de potencia:', styles['Normal'])
    #objetivo_paragraph = Paragraph(f'Función objetivo: {objetivo}', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[p6_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))


    # Obtener la ruta completa de la imagen
    img_path = "../static/facturacionPotencia.jpg"

    # Comprobar si la imagen existe y agregarla
    if os.path.exists(img_path):
        img = Image(img_path, 1.2*inch, 0.25*inch)
        elements.append(img)
        elements.append(Spacer(1, 12))
        print("Imagen agregada al PDF.")
    else:
        print(f"La imagen '{img_path}' no pudo ser encontrada.")


    # Agregar los datos del suministro
    p1_paragraph = Paragraph(f'•	FP: Facturación de la potencia.', styles['Normal'])
    p2_paragraph = Paragraph(f'•	Tpp: Precio del término de potencia del periodo horario p y año [€/kW].', styles['Normal'])
    p3_paragraph = Paragraph(f'•	Pcp: Potencia contratada en el período horario p [kW].', styles['Normal'])
    p4_paragraph = Paragraph(f'•	i: Número de periodos horarios de los que consta el término de facturación de potencia del peaje correspondiente.', styles['Normal'])
    #objetivo_paragraph = Paragraph(f'Función objetivo: {objetivo}', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[p1_paragraph], [p2_paragraph], [p3_paragraph], [p4_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))

    p6_paragraph = Paragraph(f'Para calcular los excesos de potencia:', styles['Normal'])
    #objetivo_paragraph = Paragraph(f'Función objetivo: {objetivo}', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[p6_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))

    # Obtener la ruta completa de la imagen
    img_path = "../static/excesosPotencia.jpg"

    # Comprobar si la imagen existe y agregarla
    if os.path.exists(img_path):
        img = Image(img_path, 2.3*inch, 0.3*inch)
        elements.append(img)
        elements.append(Spacer(1, 12))
        print("Imagen agregada al PDF.")
    else:
        print(f"La imagen '{img_path}' no pudo ser encontrada.")

    # Agregar los datos del suministro
    p1_paragraph = Paragraph(f'•	FEP: Facturación en concepto de excesos de potencia para equipos de medida tipo 1, 2 y 3.', styles['Normal'])
    p2_paragraph = Paragraph(f'•	KP: Relación de precios por periodo horario p, calculada como el cociente entre el término de potencia del periodo p respecto del término de potencia del periodo 1 del peaje correspondiente.', styles['Normal'])
    p3_paragraph = Paragraph(f'•	tep: Término de exceso de potencia del peaje correspondiente [€/kW].', styles['Normal'])
    p4_paragraph = Paragraph(f'•	Pdj: Potencia demandada en cada uno de los cuartos de hora j del período horario p en que se haya sobrepasado Pcp [kW]. Si el equipo de medida no dispone del registro cuartohorario, se considerará la misma potencia demandada en todos los cuartos de hora.', styles['Normal'])
    p5_paragraph = Paragraph(f'•	Pcp: Potencia contratada en el período horario p [kW].', styles['Normal'])
    p6_paragraph = Paragraph(f'•	i: Número de periodos horarios de los que consta el término de facturación de potencia del peaje correspondiente.', styles['Normal'])
    #objetivo_paragraph = Paragraph(f'Función objetivo: {objetivo}', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[p1_paragraph], [p2_paragraph], [p3_paragraph], [p4_paragraph], [p5_paragraph], [p6_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))




    # Obtener la ruta completa de la imagen
    img_path = "../static/excesosPotencia2.jpg"

    # Comprobar si la imagen existe y agregarla
    if os.path.exists(img_path):
        img = Image(img_path, 1.8*inch, 0.23*inch)
        elements.append(img)
        elements.append(Spacer(1, 12))
        print("Imagen agregada al PDF.")
    else:
        print(f"La imagen '{img_path}' no pudo ser encontrada.")
    
        # Agregar los datos del suministro
    p1_paragraph = Paragraph(f'•	FEP: Facturación en concepto de excesos de potencia para equipos de medida tipo 4 y 5.', styles['Normal'])
    p2_paragraph = Paragraph(f'•	tp: Término de exceso de potencia del peaje correspondiente [€/kW].', styles['Normal'])
    p3_paragraph = Paragraph(f'•	Pdj: Potencia demandada en cada uno de los períodos horario p en que se haya sobrepasado Pcp [kW].', styles['Normal'])
    p4_paragraph = Paragraph(f'•	Pcp: Potencia contratada en el período horario p [kW].', styles['Normal'])
    p5_paragraph = Paragraph(f'•	i: Número de periodos horarios de los que consta el término de facturación de potencia del peaje correspondiente.', styles['Normal'])
    #objetivo_paragraph = Paragraph(f'Función objetivo: {objetivo}', styles['Normal'])

    # Crear una tabla para alinear los datos del suministro
    supply_data_table = Table(
        [[p1_paragraph], [p2_paragraph], [p3_paragraph], [p4_paragraph], [p5_paragraph]],
        colWidths=[7*inch]
    )
    supply_data_table.setStyle(TableStyle([
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE')
    ]))
    elements.append(supply_data_table)
    elements.append(Spacer(1, 12))


    # Construir el PDF
    doc.build(elements)
    print(f"PDF '{filename}' creado con éxito.")

if __name__ == "__main__":
    texto = "¡Hola! Este es un PDF creado con Python y reportlab."
    imagen = "grafica_original.png"
    archivo_csv = "datosProcesadosModificados.csv"  # Asegúrate de tener este archivo en el directorio adecuado
    create_pdf("ejemplo_con_imagen_y_tabla.pdf", texto, imagen, archivo_csv)
```
